```python
import pandas as pd
import numpy as np

num_samples = 10000

countries = ['미국', '일본', '중국', '프랑스', '독일', '영국', '호주', '캐나다', '인도', '필리핀']

visa_types = ['E-7', 'F-2', 'F-5', 'D-2']
employment_types = ['정규직', '계약직', '자영업', '프리랜서']
credit_rating = [0, 1, 2, 3, 4]

data = {
    "Country": np.random.choice(countries, num_samples),
    "Residence Duration": np.random.randint(1, 20, num_samples),  
    "Visa Type": np.random.choice(visa_types, num_samples),
    "Age": np.random.randint(20, 65, num_samples),
    "Annual Income": np.random.randint(2000, 10000, num_samples),
    "Employment Type": np.random.choice(employment_types, num_samples),
    "Job Duration": np.random.randint(1, 15, num_samples),
    "Banking Duration": np.random.randint(1, 20, num_samples),
    "Avg Bank Balance": np.random.randint(50, 5000, num_samples),
    "Total Loan Amount": np.random.randint(0, 10000, num_samples),
    "Loan Ratio": np.round(np.random.uniform(0, 1, num_samples), 2),
    "Credit Card Ownership": np.random.choice([0, 1], num_samples),
    "Avg Card Spending": np.random.randint(0, 1000, num_samples),
    "Utility Payment History": np.random.choice([0, 1], num_samples),
    "Telecom Payment History": np.random.choice([0, 1], num_samples),
    "Delinquencies": np.random.randint(0, 10, num_samples),
    "Max Delinquency": np.random.randint(0, 180, num_samples),
    "Real Estate Ownership": np.random.choice([0, 1], num_samples),
    "Total Savings": np.random.randint(0, 10000, num_samples),
    "Credit Rating": np.random.choice(credit_rating, num_samples)
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 결과 샘플 5개 확인
df
```

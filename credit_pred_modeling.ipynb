```python
df_convert = df.assign(country_type = np.where(df['Country'] == '미국', 0,
                                    np.where(df['Country'] == '호주', 1,
                                    np.where(df['Country'] == '일본', 2,
                                    np.where(df['Country'] == '프랑스', 3,
                                    np.where(df['Country'] == '필리핀', 4,
                                    np.where(df['Country'] == '캐나다', 5,
                                    np.where(df['Country'] == '중국', 6,
                                    np.where(df['Country'] == '영국', 7,
                                    np.where(df['Country'] == '독일', 8, 9))))))))))
```

```python
df_convert = df_convert.assign(visa_type = np.where(df_convert['Visa Type'] == 'F-2', 0,
                                            np.where(df_convert['Visa Type'] == 'E-7', 1,
                                            np.where(df_convert['Visa Type'] == 'D-2', 2, 3))))
```

```python
df_convert = df_convert.assign(emp_type = np.where(df_convert['Employment Type'] == '프리랜서', 0,
                                            np.where(df_convert['Employment Type'] == '자영업', 1,
                                            np.where(df_convert['Employment Type'] == '정규직', 2, 3))))
```

```python
df_n = pd.read_csv('./data_zip/foreign_labor_credit_predict.csv')
df_n
```

```python
df_n.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
target = df_n['Credit Rating']
```

```python
scaling_list = ['Residence Duration', 'Age', 'Annual Income', 'Job Duration', 'Banking Duration', 'Avg Bank Balance', 'Total Loan Amount',
       'Loan Ratio', 'Total Savings']
```

```python
ss = StandardScaler()
scaling_df = df[scaling_list]
scaled = ss.fit_transform(scaling_df)
scaled_df = pd.DataFrame(scaled, columns = scaling_list)
scaled_df
```

```python
non_scaling_list = ['Credit Card Ownership', 'Avg Card Spending',
       'Utility Payment History', 'Telecom Payment History', 'Delinquencies',
       'Max Delinquency', 'Real Estate Ownership', 'country_type', 'visa_type', 'emp_type']

non_scaled_df = df_n[non_scaling_list]
non_scaled_df
```

```python
df_train = pd.concat([scaled_df, non_scaled_df], axis = 1)
```

```python
df_train
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_train, target, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
model = rf.fit(x_train, y_train)
```

```python
print("테스트 세트 정확도: {:.3f}".format(model.score(x_test, y_test)) )
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(n_estimators = 400, learning_rate = 0.05, max_depth = 3)
xgb.fit(x_train, y_train)
preds = xgb.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, preds))
```
